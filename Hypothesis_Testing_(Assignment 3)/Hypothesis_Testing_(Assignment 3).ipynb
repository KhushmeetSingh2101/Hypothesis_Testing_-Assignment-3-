```python
Cutlets = pd.read_csv('Cutlets.csv')
Cutlets.head()
```

Null Hypothesis : There is no significant difference between the diameter of the two cutlets , one each from both the Units. I.e. diameter in Unit A is Similar to diameter in Unit B.

Alternate Hypothesis : There is significant difference between the diameter of the two cutlets , one each from both the Units. I.e. diameter in Unit A is not Similar to diameter in Unit B.

```python
Cutlets = Cutlets.rename({'Unit A': 'Unit_A', 'Unit B': 'Unit_B'},axis=1)
```

```python
Cutlets.head()
```

```python
sns.pairplot(Cutlets)
```

```python
Cutlets.corr()
```

**Performing two sample test because we have two different samples to compare.**

```python
stats.ttest_ind(Cutlets['Unit_A'],Cutlets['Unit_B'])
```

```python
stats.ttest_ind(Cutlets['Unit_A'],Cutlets['Unit_B'])[1]
```

Since p-value > 0.05, then we will go with **Null Hypothesis**

It can be said that there is no significant difference in the diameter of the Cutlets between two units.

```python
LabTat = pd.read_csv('LabTAT.csv')
LabTat.head(10)
```

Null Hypothesis - There is no significant difference in the average turn around time(TAT) of reports of laboratories

Alternate Hypothesis - There is a difference in the average turn around time(TAT) of reports of laboratories



```python
LabTat = LabTat.rename({'Laboratory 1': 'Lab_1', 'Laboratory 2': 'Lab_2', 'Laboratory 3': 'Lab_3', 'Laboratory 4': 'Lab_4'}, axis=1)
```

```python
LabTat.head()
```

```python
sns.pairplot(LabTat)
```

```python
LabTat.corr()
```

```python
stats.f_oneway(LabTat['Lab_1'], LabTat['Lab_2'], LabTat['Lab_3'], LabTat['Lab_4'])
```

```python
stats.f_oneway(LabTat['Lab_1'], LabTat['Lab_2'], LabTat['Lab_3'], LabTat['Lab_4'])[1]
```

Since p-value < 0.05, then we will go with **Alternate Hypothesis**.

This means there is a difference in average TAT among the different laboratories.

```python
Customer_OF = pd.read_csv('Costomer+OrderForm.csv')
Customer_OF.head(10)
```

Null Hypothesis :- Customer order forms defective %(Percentage) does not varies by centre

Alternate Hypothesis :- Customer order forms defective %(Percentage) varies by centre

```python
print(Customer_OF['Phillippines'].value_counts())
print(Customer_OF['Indonesia'].value_counts())
print(Customer_OF['Malta'].value_counts())
print(Customer_OF['India'].value_counts())
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
Customer_OF['Phillippines']=le.fit_transform(Customer_OF['Phillippines'])
Customer_OF['Indonesia']=le.fit_transform(Customer_OF['Indonesia'])
Customer_OF['Malta']=le.fit_transform(Customer_OF['Malta'])
Customer_OF['India']=le.fit_transform(Customer_OF['India'])
```

```python
Customer_OF
```

```python
fig = plt.figure(figsize=(5,3))
plt.xlabel('Phillippines')
plt.ylabel('Number of 0s & 1s')
plt.hist(Customer_OF['Phillippines'])
```

```python
fig = plt.figure(figsize=(5,3))
plt.xlabel('Indonesia')
plt.ylabel('Number of 0s & 1s')
plt.hist(Customer_OF['Indonesia'])
```

```python
fig = plt.figure(figsize=(5,3))
plt.xlabel('Malta')
plt.ylabel('Number of 0s & 1s')
plt.hist(Customer_OF['Malta'])
```

```python
fig = plt.figure(figsize=(5,3))
plt.xlabel('India')
plt.ylabel('Number of 0s & 1s')
plt.hist(Customer_OF['India'])
```

```python
stats.f_oneway(Customer_OF['Phillippines'], Customer_OF['Indonesia'], Customer_OF['Malta'], Customer_OF['India'])
```

Since p-value > 0.05, then we will go with Null Hypothesis

It can be said that customer order forms defective % does not varies by centre.
